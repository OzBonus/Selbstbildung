# Big O Notation

Big O notation is a method to communicate how the running time of an algorithm varies as a function of its input. It's strength is in being generalizable to any computer.

I've been through a lot of tutorials, so I'll skip the majproty of the notes.

A few things I want to run through in my mind:

* We are interested in comparing how quickly runtime grows, rather than comparing actual runtimes.

* As n gets arbitrarily large, only those parts of the algorithm that most greatly affect the runtime are taken into account. Other parts are ignored.

# Common Big O Runtimes

| Big O | Name | Acceptability |
| ----- | ---- | ------------- |
| O(1) | Constant | <font color='green'>**Great!**</font>
| O(log n) | Logarithmic | <font color='green'>**Great!**</font>
| O(n) | Linear | <font color='orange'>**Fine**</font>
| O(n log n) | Log Linear | <font color='orange'>**Fine**</font>
| O(n<sup>2</sup>) | Quadratic | <font color='orangered'>**Bad**</font>
| O(n<sup>3</sup>) | Cubic | <font color='orangered'>**Bad**</font>
| O(2<sup>n</sup>) | Exponential | <font color='darkred'>**Danger!**</font>
| O(n!) | Factorial | <font color='darkred'>**Danger!**</font>

# Graphing Runtimes

This is going to be fun. That wasn't sarcastic.

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('bmh')

# Prepare runtime comparisons.
# Make an array of values for n.
# np.linspace(1, 20, 100) would be smoother, but np.arange(1, 21) is more realistic.
# Also, np.arange is required for plotting factorials of n because integers are required.
n = np.arange(1, 21)
labels = ['Constant', 'Logarithmic', 'Linear', 'Log Linear', 'Quadratic', 'Cubic', 'Exponential', 'Factorial']
big_o = [np.ones(n.shape), np.log(n), n, n*np.log(n), n**2, n**3, 2**n, list(map(lambda x: np.math.factorial(x), n))]
# Change to ones.^         ^          ^  ^            ^     ^     ^     ^
# log10(n). ---------------^          ^  ^            ^     ^     ^     ^
# Just return n[i]. ------------------^  ^            ^     ^     ^     ^
# n to the power of log10(n). -----------^            ^     ^     ^     ^
# n squared. -----------------------------------------^     ^     ^     ^
# n cubed. -------------------------------------------------^     ^     ^
# 2 to the power of n. -------------------------------------------^     ^
# n factorial. ---------------------------------------------------------^

# Prepare the plot.
plt.figure(figsize=(20, 20))
plt.ylim(0, 1000)
plt.xticks(np.arange(min(n), max(n)+1, 1))

for i in range(len(big_o)):    
    plt.plot(n, big_o[i], label = labels[i])

plt.legend(loc=0)
plt.ylabel('Relative Runtime')
plt.xlabel('n')
```
