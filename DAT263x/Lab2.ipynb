## Perform a basic frequency analysis.

#### Obtain a text document.

```python
!curl https://raw.githubusercontent.com/MicrosoftLearning/AI-Introduction/master/files/Moon.txt -o Moon.txt

with open('Moon.txt', 'r') as stream:
    txt_1 = stream.read()
print(txt_1)
```

#### Normalize the text.

```python
txt_1 = ''.join(char for char in txt_1 if char.isalpha() or char.isspace()).lower()
print(txt_1)
```

#### Create a frequency distribution of words.

```python
import nltk
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from nltk.probability import FreqDist
nltk.download('punkt')

tokens = nltk.tokenize.word_tokenize(txt_1)
dist = FreqDist(tokens)
count_df = pd.DataFrame(dist, index=[0]).transpose()
count_df.columns = ['count']
count_df.head()
```

#### Make a Pareto chart of the most common words.

```python
fig, ax = plt.subplots(figsize=(15, 5), facecolor='w')
count_df.sort_values(by='count', ascending=False)[:60].plot(kind='bar', ax=ax)
ax.set(xlabel='Word', ylabel='Frequency', title='Frequency of the Most Common Words')
plt.show()
```

#### Remove stop words and visualize again.

```python
from nltk.corpus import stopwords
```

```python
stopwords_list = stopwords.words('english')

txt_1 = ' '.join([word for word in txt_1.split() if word not in stopwords_list])
tokens = nltk.tokenize.word_tokenize(txt_1)
dist = FreqDist(tokens)
count_df = pd.DataFrame(dist, index=[0]).transpose()
count_df.columns = ['count']

fig, ax = plt.subplots(figsize=(15, 5), facecolor='w')
count_df.sort_values(by='count', ascending=False)[:60].plot(kind='bar', ax=ax)
ax.set(xlabel='Word', ylabel='Frequency', title='Frequency of the Most Common Words, Stop Words Excluded')
plt.show()
```

## Perform a *term frequency, inverse document frequency* analysis.

When dealing with a corpus coprising many documents, a more commonly used technique, rather than basic frequency analysis is *term frequency, inverse document frequency* (TF-IDF). This is a score calculated by comparing how often a word or phrase appears in one document with how often it appears appears across a collection of documents. The metric assigns a high level of relevance to words or phrases that appear frequently in a document, but less frequently across a range of documents. This would be useful to discern, for example, what the topic of an article might be.

#### Load and prepare several text documents.

```python
!curl https://raw.githubusercontent.com/MicrosoftLearning/AI-Introduction/master/files/Moon.txt -o Moon.txt
!curl https://raw.githubusercontent.com/MicrosoftLearning/AI-Introduction/master/files/Gettysburg.txt -o Gettysburg.txt
!curl https://raw.githubusercontent.com/MicrosoftLearning/AI-Introduction/master/files/Cognitive.txt -o Cognitive.txt
!curl https://raw.githubusercontent.com/MicrosoftLearning/AI-Introduction/master/files/Inaugural.txt -o Inaugural.txt
!iconv -f utf-16  -t us-ascii -o Inaugural.txt Inaugural.txt
    
corpus = ['Moon.txt', 'Gettysburg.txt', 'Cognitive.txt', 'Inaugural.txt']
```

```python
from textblob import TextBlob as tb

def make_blob(file):
    with open(file, 'r') as stream:
        text = stream.read().lower().split()
        text = ' '.join(word for word in text if word not in stopwords_list and not word.isnumeric())
    return tb(text)

text1, text2, text3, text4 = (make_blob(doc) for doc in corpus)
```

#### Define functions.

```python
import math

bloblist = [text1, text2, text3, text4]

def tf(word, blob):
    """Calculate term frequency."""
    return blob.words.count(word) / len(blob.words)

def n_containing(word, bloblist):
    """The number of docs that contain word."""
    return sum(1 for blob in bloblist if word in blob.words)

def idf(word, bloblist):
    """Calculate inverse document frequency."""
    return math.log(len(bloblist)) / (1 + n_containing(word, bloblist))

def tfidf(word, blob, bloblist):
    """Calculate term frequency-inverse document frequency"""
    return tf(word, blob) * idf(word, bloblist)
```

#### Calculate TF-IDF.

```python
for i, blob in enumerate(bloblist, start=1):
    print(f'The top words in text{i} are:')
    scores = {word: tfidf(word, blob, bloblist) for word in blob.words}
    sorted_words = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    for word, score in sorted_words[:5]:
        print(f'\t{word:<13} TF-IDF: {round(score, 5)}')
```

## Consider word stems in an analysis.

The analysis above would treat *nation* and *national* as two wholly different words. We can better results if we consider only the base words in a document.

#### Make a simple Pareto chart of the words in Kennedy's inaugural address.

```python
dist = FreqDist(text4.words)
kennedy_df = pd.DataFrame(dist, index=[0]).transpose()
kennedy_df.columns = ['count']

fig, ax = plt.subplots(figsize=(15, 5), facecolor='white')
kennedy_df.sort_values(by='count', ascending=False)[:60].plot(kind='bar', ax=ax)
ax.set(xlabel='Frequency', ylabel='Word', title='Frequency of the Most Common Words in Kenndey\'s Inauguration Speech')
plt.show()
```

#### Stem the words and plot again.

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
stems = [ps.stem(word) for word in text4.words]


stem_df = pd.DataFrame(FreqDist(stems), index=[0]).transpose()
stem_df.columns = ['count']

fig, ax = plt.subplots(figsize=(15, 5), facecolor='white')
stem_df.sort_values(by='count', ascending=False)[:60].plot(kind='bar', ax=ax)
ax.set(xlabel='Frequency', ylabel='Word', title='Frequency of the Most Common Word Stems in Kenndey\'s Inauguration Speech')
plt.show()
```

Notice how *power* and *nation* have leapt forwards in the rankings!.
