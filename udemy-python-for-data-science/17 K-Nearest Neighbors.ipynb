## Standard Scores

In order to comapre values between different distributions, we must transform each value into a point on a single, standard scale.. Otherwise, it would be difficult to compare values that are distributed over very different scales, such as -1 to 1 vs 0 to 100. For KNN in particular, points on a larger scale could be considered further away than points on a smaller scale.

I don't know if sklearn uses this method, but one that I learned at uni was $z$-score.

$$
\large
z = \frac{\mu - \sigma}{\sigma}
$$

where $\mu$ is the mean and $\sigma$ is the standard deviation.

Scikit-Learn includes scaling tools in its preprocessing submodule.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(anon.drop('TARGET CLASS', axis=1)) # Compute the mean and stdev for later scaling.
scaled_features = scaler.transform(anon.drop('TARGET CLASS', axis=1)) # Perform standardization by centering and scaling.
scaled_anon = pd.DataFrame(scaled_features, columns=anon.columns[:-1])
scaled_anon.head()
```

## Train Test Split

Developing a KNN model requires a train-test split, just like any other.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(scaled_anon, anon['TARGET CLASS'], test_size=0.3)
```

## Training and Prediction

Just for educational purposes, we will start with $k=1$.

```python
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors=1)
classifier.fit(X_train, y_train)
predictions = classifier.predict(X_test)
```

## Assessment

So how did our model perform?

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
```

Actually, not too bad. Probably this is thanks to the manner in which the data were randomly generated.

## Choosing a Better Value for k

In a real-world scenario, $k=1$ is unlikely to produce to great results, but so is using a too-large value for $k$. So, how do we decide which value to use? Let's use the elbow method, which is, as taken from Wikipedia:

> This method looks at the percentage of variance explained as a function of the number of clusters: One should choose a number of clusters so that adding another cluster doesn't give much better modeling of the data.

So, for a value for $k$ which we will call $n$, if $n+1$ can account for significantly more variance than $n$, then we go with $n+1$. Then we repeat the process until adding to $n$ no longer succeeds in accounting for much variance. That might not be the most eloquent description, but I hope you get it.

Let's demonstrate this visually.

```python
error_rates = []
max_k = 50

# This might take a while to execute.
for i in range(1, max_k):
    classifier = KNeighborsClassifier(n_neighbors=i)
    classifier.fit(X_train, y_train)
    predictions = classifier.predict(X_test)
    error_rates.append(np.mean(predictions != y_test))

fig, ax = plt.subplots()
ax.plot(range(1, max_k),
        error_rates,
        color='blue',
        linestyle='dashed',
        marker='o',
        markerfacecolor='orange',
        markersize=10)
ax.set_title('Error Rates for Values of k')
ax.set_xlabel('k')
ax.set_ylabel('Error Rate')
ax.grid(False)
plt.show()
```

Just by eyeballing it, $k=30$ looks like a good enough choice, though I suppose someone might have developed a more formal selection method.

Let's compare $k=1$ and $k=30$ side by side, shall we?

```python
def do_knn(k):
    classifier = KNeighborsClassifier(n_neighbors=k)
    classifier.fit(X_train, y_train)
    predictions = classifier.predict(X_test)
    print(confusion_matrix(y_test, predictions))
    print(classification_report(y_test, predictions))    
```

```python
do_knn(1)
```

```python
do_knn(30)
```

Using $k=30$ does indeed lead to a much more precise model.
