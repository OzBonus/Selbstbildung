# K Means Clustering

K Means Clustering is an unsupervised learning algorithm that attempts to cluster data into a $k$ groups. This $k$ is predefined by the user. Here's the outline of the algorithm.

1. The algorithm begins by assigning each observation to one of the $k$ clusters at random and then finding each cluster's centroid.
1. Iterate on the following until no more changes occur:
    1. Reassign data points to the cluster of the nearest centroid.
    1. Calculate new centroids for each cluster.
    
The goal of the algorithm is to minimize within cluster variance as measured by the sum of the euclidean distance between a cluster's data points and its centroid.

## Imports

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs

%matplotlib inline
```

## Artificial Data

```python
data = make_blobs(n_samples=200,
                  n_features=2,
                  centers=4,
                  cluster_std=1.8,
                  random_state=101)
```

## Visualize

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 5), facecolor='w')
ax1.scatter(x=data[0][:,0], y=data[0][:,1])
ax1.set_title('Just the Data')

ax2.scatter(x=data[0][:,0], y=data[0][:,1], c=data[1], cmap='Accent')
ax2.set_title('True Clusters')

for ax in (ax1, ax2):
    ax.tick_params(axis='both',
                   bottom=False,
                   left=False,
                   labelbottom=False,
                   labelleft=False)

plt.tight_layout()
plt.show()
```

## Defining the Clusters

An important part of this model, what may be a weakness or a strength, is that you must define a number &k& or clusters.

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(data[0])
```

Let's take a look at where those centroids ended up.

```python
kmeans.cluster_centers_
```

We can access an array of the cluster labels.

```python
kmeans.labels_
```

Let's compare the model's results with the real clusters. In practice, we'd never get the chance to do this with data from the wild, but only with artificial data like what we've got here.

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 5), facecolor='w')

ax1.set_title('K Means')
ax1.scatter(x=data[0][:,0], y=data[0][:,1], c=kmeans.labels_, cmap='Accent')
ax1.scatter(x=kmeans.cluster_centers_[:,0],
            y=kmeans.cluster_centers_[:,1],
            s=1024, marker='+')

ax2.set_title('True Clusters')
ax2.scatter(x=data[0][:,0], y=data[0][:,1], c=data[1], cmap='Accent')

for ax in (ax1, ax2):
    ax.tick_params(axis='both',
                   bottom=False,
                   left=False,
                   labelbottom=False,
                   labelleft=False)

plt.tight_layout()
plt.show()
```

The model did a pretty good job of assigning the observations to the right clusters, though it wasn't perfect. Although this model wouldn't go far in proving anything on its own, it does offer inspiration for where to direct one's supervised models.
