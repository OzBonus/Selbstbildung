# Regression Plots with `sns.lmplot()`

This is just an overview until we get into proper linear regression techniques later on in the machine learning sections.

```python
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
tips.head()
```

The basics of calling this plot is much like any other Seaborn plot.

```python
sns.lmplot(data=tips,
           x='total_bill',
           y='tip')
```

Categorical data can be differentiated with different colors.

```python
sns.lmplot(data=tips,
           x='total_bill',
           y='tip',
           hue='sex')
```

Make categorical separation more obvious by spreading them across rows and columns.

```python
# Separate sex.
sns.lmplot(data=tips,
           x='total_bill',
           y='tip',
           col='sex')
```

```python
# Separate sex and time.
sns.lmplot(data=tips,
           x='total_bill',
           y='tip',
           col='sex',
           row='time')
```

```python
# Combine a grid layout with hue separation.
sns.lmplot(data=tips,
           x='total_bill',
           y='tip',
           col='day',
           hue='sex')
```

Complex grids may need some size adjustments. This will be expanded upon in another section.

```python
# Combine a grid layout with hue separation.
sns.lmplot(data=tips,
           x='total_bill',
           y='tip',
           col='day',
           hue='sex',
           aspect=0.5, # The aspect ratio of each subplot.
           size=8) # A scalar value, in inches. 我的天！
```
