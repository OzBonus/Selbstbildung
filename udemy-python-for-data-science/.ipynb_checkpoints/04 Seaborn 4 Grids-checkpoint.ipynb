# Grids

Seaborn makes it rather easy to toss out a bunch a subplots in a nicely arranged grid layout.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

# `sns.PairGrid()`

The `sns.PairGrid()` object allows for far more customization than `sns.pairplot()`. First, take a look at what `sns.pairplot()` prints out.

```python
sns.pairplot(iris)
```

Doing the same with `sns.PairGrid()` produces and empty plot that will be used as a canvas.

```python
sns.PairGrid(iris)
```

Assigning the grid to a variable is a good first step. Then you can map a plot type (here we are using a scatterplot as provided by matplotlib) to each subplot of the grid.

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

One way to divide a `PairGrid` is by the diagonal, the upper area, and the lower area. Seaborn makes this simple to do.

```python
# This might take a while.
g = sns.PairGrid(iris,
                 hue='species',
                 palette='rainbow')
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

Pair plots can still be cool though. And it's not like you're likely going to use a bunch a plots like these any way.

```python
sns.pairplot(data=iris,
             hue='species',
             palette='rainbow')
```
