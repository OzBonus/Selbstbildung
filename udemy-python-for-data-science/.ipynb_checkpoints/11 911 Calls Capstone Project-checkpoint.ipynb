Do the necessary imports for pandas, numpy, seaborn, and matplotlib.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

A zipped copy of records of 911 calls has already been placed in your library. In case you didn't already know, pandas can retreive files within compressed containers, so long there is just one file.

Read the file into a data frame and display the head and description.

```python
df = pd.read_csv('911.zip')

display(df.info())
display(df.head())
```

## What are the top five ZIP codes for 911 calls?

```python
df['zip'].value_counts().iloc[:5]
```

## What are the top five townships for 911 calls?

```python
df['twp'].value_counts().iloc[:5]
```

## How many title codes are there?

```python
df['title'].nunique()
```

## Create a new column `reason` from the fist part of `title`.

```python
df['reason'] = df['title'].apply(lambda v: v.split(':')[0])

df.head()
```

## What is the most common reason for a 911 call?

```python
df['reason'].value_counts()
```

## Use seaborn to make a bar chart of `reason`.

```python
sns.countplot(df['reason'])
```

## What data type is used in the `timeStamp` column?

```python
# Should return str ... but it doesn't exactly.
df['timeStamp'].dtype
```

## Convert that column into proper datetime format.

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'], infer_datetime_format=True)

df['timeStamp'].head()
```

Now that you have a column using a proper datetime format, more interesting work can be done.

## Make columns for month, day, and hour.

Your next task is to make columns that contain data representing only the month, day, or hour of the `timeStamp` column. This will make further tasks much simpler. Although it would be possible to just plug in the `timeStamp` column later on, having these row already set aside will make the colde much cleaner than rambling one-liners.

```python
# Create a month column.
df['month'] = df['timeStamp'].apply(lambda t: t.month)
df['month'].head()
```

```python
# Create an hour column.
df['hour'] = df['timeStamp'].apply(lambda t: t.hour)
```

```python
# Create a day column.
day_map = {0: 'Mon', 1: 'Tue',
           2: 'Wed', 3: 'Thu',
           4: 'Fri', 5: 'Sat',
           6: 'Sun'}

df['day'] = df['timeStamp'].apply(lambda t: day_map[t.dayofweek])
```

Now take a peek at the df to make sure that everything is in place.

```python
df.head()
```

## Make some plot finally.

Your first step into actually grokking some insight hidden in the data will be accomplished by making a seaborn countplot that counts up the number of calls on each day of the week and splits those calls by their reason.

```python
sns.countplot(data=df, x='day', hue='reason')
plt.legend(bbox_to_anchor=(1, 1), loc=2)
```

At a glance it looks like the count of calls vary as a function of the day of the week, but some types of calls appear to be more strongly affected than others. That's interesting to know.

Now repeat the above, but for the month.

```python
sns.countplot(data = df,
              x = 'month',
              hue = 'reason')
plt.legend(bbox_to_anchor=(1, 1), loc=2)
```

November and January look quite peaceful, don't they? Might that have something to do with the weather? Perhaps people are just more likely to stay inside and under a blanket during such cold months.

Now I would like you to show how many calls occured on each date in the dataset. First you should create a new column called `date` and use the `.date()` method to convert the string-formatted time stamps. After that, group the dataset by the new `date` column and plot the dates against the number of calls made.

```python
df['date'] = df['timeStamp'].apply(lambda t: t.date())
date_group = df.groupby('date', as_index=False).count()

fig, axes = plt.subplots(figsize=(15, 5))
axes.set_title('911 Calls Per Date')
axes.set_xlabel('Date')
axes.set_ylabel('Number of 911 Calls')
plt.plot(date_group['date'], date_group['e'])
```

That looks halfway decent, and you got to review a little bit of matplotlib, which makes that a double-win.

Next, make three more plots like the one above, but have each on show one of the three reasons for each call having been made. You can start by making a groupby object that has a heirachial index and use that for all three plots.

```python
dr_group = df.groupby(['date', 'reason']).count().unstack()
dr_group.head(9)
```

```python
fig, (axes1, axes2, axes3) = plt.subplots(nrows=3, ncols=1, figsize=(15, 10), sharex=True)

axes1.plot(dr_group['e']['EMS'])
axes1.set_title('EMS')

axes2.plot(dr_group['e']['Fire'])
axes2.set_title('Fire')

axes3.plot(dr_group['e']['Traffic'])
axes3.set_title('Traffic')

fig.suptitle('911 Calls by Date and Reason')
```

That looks pretty good. You had some trouble, but you got it worked out in the end. It isn't a cool-looking as you imagined, but it looks better than the three totally seperate plots used in the course notebook, so you got that going for you.

The next phase of the analysis involves making heatmaps. The first heatmap will compare the day of the week to the hour of the day. You can do this more easily if you group by day and hour, then unstack the hour. Maybe there is a better way that doesn't involve that second step, but that remains to be learned.

For now, make that unstacked groupby object.

```python
dh_group = df.groupby(['day', 'hour'])['e'].count().unstack()
dh_group
```

So you made a useful dataframe, good. You also tried a long time to get the days sorted, but you failed to find a way that worked. That's unfortunate, but no reason to give up. Keep on going and make a heatmap using seaborn.

```python
fig, ax = plt.subplots(figsize=(15, 10))
sns.heatmap(dh_group)
```

That's the way to do it! Workdays all seem to have roughly equivalent appearances, but weekends look to be shifted a few hours later, which suggests that everybody is staying in bed a bit longer and going to bed later. That includes going to bed late on Sunday night, which is just irresponsible since people have got work and school the next day. Also, taking a second look at Frdiday, it appears 911 calls continue later into the evening than on other working days, suggesting that people are getting an early start on their reckless weekends.

Try creating a cluster map to further highlight differences.

```python
sns.clustermap(dh_group)
```

Brilliant! This kind of information would be useful. For example, you can see what days and times would benefit from having more operators and emergency services available.

The next step is to repeat the above plots but for month plotted against the day of the week. Could you get two plots in one figure? Could you do everything in just one cell? Let's see what you're capable of!

EDIT: OK, clustermaps don't like to be plotted in the same figure as other plots.

```python
md_group = df.groupby(['month', 'day'])['e'].sum().unstack()

# plt.show() is necessary to show multiple figures in one cell.
sns.heatmap(md_group)
plt.show()

sns.clustermap(md_group)
plt.show()
```

November is a remarkably quiet month. Maybe it has something to do with the weather and the nature of the holiday season (in America, where this data is from). Sundays get a little bit more wild during times when schools are more likely to be on holiday (spring break occurs in April mostly, I believe).
