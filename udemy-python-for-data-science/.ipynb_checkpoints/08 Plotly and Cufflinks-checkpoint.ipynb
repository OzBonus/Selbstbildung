```python
import numpy as np
import pandas as pd
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import cufflinks as cf

%matplotlib inline

init_notebook_mode(connected=True)
cf.go_offline()
```

```python
df1 = pd.DataFrame(np.random.randn(100, 4),
                  columns='A B C D'.split())

df2 = pd.DataFrame({'Category':['A', 'B', 'C', 'D'], 'Values':[12,90,34,78]})

display(df1.head())
display(df2.head())
```

## Cufflinks and `df.iplot()`

Cufflinks brings the `iplot()` method to pandas, which allows integration with Plotly plots, even when offline. That's pretty cool. Though I guess the real charm with Plotly is online dashboards. Hey, this might actually get me a job somehow -- cool!

## `kind='scatter'`

```python
df.iplot(kind='scatter',
         x='A',
         y='B',
         mode='markers',
         size=10)
```

## `kind='bar'`

Plotly's default bar plot uses a sum function.

```python
df2.iplot(kind='bar',
          x='Category',
          y='Values')
```

You can put in your own grouping function, like count.

```python
df1.count().iplot(kind='bar')
```

```python
df1.mean().iplot(kind='bar')
```

## `kind='box'`

```python
df1.iplot(kind='box')
```

## `kind='surface'`

```python
df3 = pd.DataFrame({'x':[1, 2, 3],
                    'y':[3, 2, 1],
                    'z':[1, 2, 3]})
df3.iplot(kind='surface', colorscale='rdylbu')
```

## `kind='spread'`

```python
df1[['A', 'B']].iplot(kind='spread')
```

## `kind='hist'`

```python
df1['A'].iplot(kind='hist')
```

```python
df.iplot(kind='hist')
```

## `kind='bubble'`

```python
df1.iplot(kind='bubble',
          x='A',
          y='B',
          size='C')
```

## `df.scatter_matrix()`

This is very similar to Seaborn's pairplots.
