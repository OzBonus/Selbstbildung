| Feature | Description |
| ------- | ----------- |

## Time

The `Timestamp` column could be valuable. It actually probably isn't, because this is a slapdash attempt at making a fake dataset, but let's pretend that we've got something potentially useful. As-is, we can't use this column for logistic regression, but how can we change it to make it more useful? We'll somehow to need to carve it into discrete chunks of time, but which ones? We could use hours, days, weekends, months.

Well, let's just check out what the minimum and maximum values are.

```python
data['Timestamp'].min()
```

```python
data['Timestamp'].max()
```

So, the first seven months of 2016, minus a week. I feel like this gives a little insight into when this dataset was made.

Were this a real dataset, I would consider simply dividing the timestamps into typical working and non-working times. However, no timezone data is included, which means everything is based on the local timezone of whoever created this dataset, which was probably somewhere in America, but we can't know for sure. Even if we did know, we'd still have to figure out what timezone each user was in, so fuck that for obvious reasons. In conclusion, the hourly data is useless.

But how about days of the week? This is still affected by the timezone issue, but not as much. So, we will make a new column `Weekend` that codes for whether the timestamp was made during a weekend (`1`) or not (`0`).

First, let's turn the `Timestamp` column into datetime objects that pandas can work with.

```python
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
```

Now, for the conversion.

```python
data['Weekend'] = data['Timestamp'].apply(lambda x: 0 if x.weekday() < 5 else 1)
```

Take a peek inside to make sure it looks OK.

```python
data.head(10)
```

## Logistic Regression

Begin by splitting the data into training and testing chunks.

```python
X = data.drop(['Clicked on Ad', 'Timestamp'], axis=1)
y = data['Clicked on Ad']

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

Train the model with the training chunks.

```python
logmodel = LogisticRegression()
logmodel.fit(X_train, y_train)
```

```python
pd.DataFrame(logmodel.coef_, columns=X_train.columns, index=['Coefficient'])
```

So, given this it looks like the age is positively correlated with the probability of clicking on an advertisement and this relationship is far stronger than any other. The next strongest is the amount of time spent on the website, which has a relatively weak negative correlation. Everything else looks like noise.

Anyway, let's move onto the predictions phase.

```python
predictions = logmodel.predict(X_test)

print(sklearn.metrics.classification_report(y_test, predictions))
```
