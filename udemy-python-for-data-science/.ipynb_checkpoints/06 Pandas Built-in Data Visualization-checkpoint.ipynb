# Pandas Built-in Data Visualization

Pandas is bundled with the ability to make a couple of basic plots from its dataframes. Although libraries like Seaborn are preferred for making presentation-quality visualizations, the ones that come with pandas are idel for quick overviews and sanity checks.

## Do the necessary imports.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Import the sample data.

```python
df1 = pd.read_csv('df1.csv', index_col=0)
df2 = pd.read_csv('df2.csv')
df3 = pd.read_csv('df3.csv')

display(df1.head())
display(df2.head())
display(df3.head())
```

## `df.plot.hist()`

There is more than one way to call a plot through pandas. In the next few examples you will be using the format in the title. You also could have used `df.plot(type='hist')` or even `df.hist()`. That there are several different ways to do this feel unpythonic.

```python
df1['A'].plot.hist()
```

```python
df1['B'].plot.hist(bins=20)
```

## Styling your plots.

Matplotlib, which is at the core of all these plots, allows for the use of style sheets that broadly change the appearance of plots. Simply running the line `plt.style.use(*kwargs)` is enough to change the style of every following plot.

What follows is a demonstration of some styles.

```python
plt.style.use('default')
df1['C'].plot.hist()
```

```python
plt.style.use('ggplot')
df1['C'].plot.hist()
```

```python
plt.style.use('bmh')
df1['C'].plot.hist()
```

```python
plt.style.use('dark_background')
df1['C'].plot.hist()
```

```python
plt.style.use('fivethirtyeight')
df1['C'].plot.hist()
```

```python
plt.style.use('default')
```

## `df.plot.area()`

```python
df2.plot.area()
```

## `df.plot.bar()`

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

## `df.plot.line()`

```python
df1.plot.line(x=df1.index,
              y='B',
              figsize=(12, 3),
              lw=1)
```

# `df.plot.scatter()`

```python
df1.plot.scatter(x='A', y='B')
```
