# Basic Recommender Systems

## Imports

```python
import numpy as np
import pandas as pd
```

## Data

```python
df = pd.read_csv('u.data', sep='\t', names=['user_id', 'item_id', 'rating', 'timestamp'])
df.head()
```

```python
titles = pd.read_csv('Movie_Id_Titles')
titles.head()
```

```python
df = pd.merge(df, titles, on='item_id')
df.head()
```

## Poking Around

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
%matplotlib inline
```

How many ratings does each film have? Showing them all at once isn't practical, but we can get an idea by sorting them by the sum of their reviews and then showing every 50th movie.

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].sum().sort_values(ascending=False))
ratings.columns = ['n_ratings']
ratings.iloc[::50]
```

The count of ratings varies quite a bit, as you can see. How about we plot these counts with a histogram?

```python
fig, ax = plt.subplots(figsize=(15, 10))
sns.distplot(ratings['n_ratings'], kde=False)
ax.set_xlim(0)
plt.show()
```

An overwhelming number of movies have very few ratings.

Next, what is the average rating for each film? Let's just look at every 50th film like we did before.

```python
ratings['avg_rating'] = df.groupby('title')['rating'].mean() # Pandas will match the indicies.
ratings.sort_values(by='avg_rating', ascending=False)[::50]
```

Movies with integer ratings like `4.000000` indicate that they received a rating from only one person. Let's graph this as well.

```python
fig, ax = plt.subplots(figsize=(15, 10))
sns.distplot(ratings['avg_rating'], kde=False, bins=50)
ax.set_xlim(1, 5)
plt.show()
```

You can clearly see that there are lot of movies with just a single rating of 1, 2, 3, 4, or 5 stars. Aside from those the distribution looks relatively normal.

How does a movie's rating compare to the number of ratings that is has? We can see this with a joint plot.

```python
sns.jointplot(data=ratings, x='avg_rating', y='n_ratings', alpha=0.2)
```

Cool looking graph, isn't it? It shouldn't be a great surprise that there is a relationship between popularity (by proxy of the number of ratings) and average rating.

## Recommending Similar Movies

We'll begin by creating a pivot table whereing the each user is a row and their ratings for each movie are the columns. There will be many `NaN` values because most users will not have rated most movies.

```python
movie_matrix = df.pivot_table(index='user_id', columns='title', values='rating')
movie_matrix.head()
```

Let's find movies similar to *Star Wars*. Pandas `DataFrame.corrwith()` method can compare the pairwise similarity between objects.

```python
starwars_ratings = movie_matrix['Star Wars (1977)']
```

```python
similar_to_starwars = pd.DataFrame(movie_matrix.corrwith(starwars_ratings), columns=['corr'])
similar_to_starwars.dropna(inplace=True)
similar_to_starwars.sort_values(ascending=False, by='corr').head(10)
```

Well, these results look strange. Probably because some movies with very few ratings are getting bumped up higher than they should be. We can deal with this by limiting the correlation to films with a certain number of ratings.

```python
enough_ratings = movie_matrix.columns[movie_matrix.apply(lambda col: col.notnull().sum() >= 100)]
limited_movie_matrix = movie_matrix[enough_ratings]

starwars_ratings = movie_matrix['Star Wars (1977)']
similar_to_starwars = pd.DataFrame(limited_movie_matrix.corrwith(starwars_ratings), columns=['corr'])
similar_to_starwars.dropna(inplace=True)
similar_to_starwars.sort_values(ascending=False, by='corr').head(20)
```

That looks much better! What movies are similar to *Titanic*?

```python
titanic_ratings = movie_matrix['Titanic (1997)']
similar_to_titanic = pd.DataFrame(limited_movie_matrix.corrwith(titanic_ratings), columns=['corr'])
similar_to_titanic.dropna(inplace=True)
similar_to_titanic.sort_values(ascending=False, by='corr').head(20)
```

And how about *Pocahontas*?

```python
pocahontas_ratings = movie_matrix['Pocahontas (1995)']
similar_to_pocahontas = pd.DataFrame(limited_movie_matrix.corrwith(pocahontas_ratings), columns=['corr'])
similar_to_pocahontas.dropna(inplace=True)
similar_to_pocahontas.sort_values(ascending=False, by='corr').head(20)
```
