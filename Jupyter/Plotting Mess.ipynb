Necessary imports

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
import string
```

These generic lists might be useful.

```python
ten = list(range(1, 11))
hundred = list(range(1, 101))
thousand = list(range(1, 1001))
twentieth = list(range(1900, 2000))
alphabet = list(string.ascii_lowercase)

rand_10 = sorted(random.sample(range(10000), 10))
rand_100 = sorted(random.sample(range(10000), 100))
rand_1000 = sorted(random.sample(range(10000), 1000))
```

Let's make sure these lists are doing what I expect them to.

```python
for item in (ten,
             hundred,
             thousand,
             twentieth,
             alphabet,
             rand_10,
             rand_100,
             rand_1000):
    print(item[:5])
```
