# Visualizing Data with matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Only for notebooks.
%matplotlib inline
```

### Make some sample data to plot.

```python
import numpy as np
x = np.linspace(0, 8, 13)
y = x**2

x[:3], y[:3]
```

### Make a simple plot.

```python
plt.plot(x, y, 'r') # 'r' denotes a solid red line.
plt.xlabel('This is the x-axis.')
plt.ylabel('This is the y-axis.')
plt.title('This is the title.')
plt.show()
```

### Try the object-oriented API.

This still seems weird.

```python
# Create Figure (empty canvas)
# There is a Figure class, but it is not called directly.
# plt.figure is a maker function that create a Figure instance.
fig = plt.figure()

# Add set of axes to figure
# fig.add_axes([left, bottom, width, height]) from 0 to 1.
# left and bottom have no effect on single plots, only mulitiplots.
axes = fig.add_axes([1, 1, 1, 1])

# Plot on that set of axes
axes.plot(x, y, 'b--')

# Set the labels.
axes.set_xlabel('Awesome x-axis')
axes.set_ylabel('Awesome y-axis')
axes.set_title('Awesome Title')

plt.show()
```

This code looks better ot my eyes, but I'm still not in love with it.

```python
fig, axes = plt.subplots(figsize=(6, 6))

axes.plot(x, y)

axes.set_title('Learning to Plot')
axes.set_xlabel('Awesome x-axis')
axes.set_ylabel('Awesome y-axis')

plt.show()
```

### Add another plot within.

The object-oriented method makes more sense when there are multiple plots to show.

```python
fig = plt.figure(figsize=(6,6))

axes1 = fig.add_axes([0.0, 0.0, 1.0, 1.0]) # Bottom left, full size of the figure.
axes2 = fig.add_axes([0.1, 0.7, 0.2, 0.2]) # Upper left, 20% of the figure's size.

axes1.plot(x, y, 's')
axes1.set(xlabel='Awesome x-axis 1', ylabel='Awesome y-axis 1', title='Awesome Title 1')

axes2.plot(x, y, 'x')
axes2.set(xlabel='Awesome x-axis 2', ylabel='Awesome y-axis 2', title='Awesome Title 2')

plt.show()
```

### Make side by side subplots.

Plots on top of other plots are perhaps not super common (although I bet you can make some cool effecs with image overlays). `plt.subplots()` is a more automated approach to subplots based on grids.

```python
fig, (axes1, axes2) = plt.subplots(nrows=1, ncols=2)

axes1.plot(x, y)
axes1.set(xlabel='x', ylabel='y', title='Left Plot')

axes2.plot(y, x)
axes2.set(xlabel='x', ylabel='y', title='Right Plot')

fig.suptitle('Look at my plots. My plots are amazing.')
plt.show()
```

### Add more lines and a legend.

```python
x = np.linspace(0, 20, 1000)

fig = plt.figure(figsize=(10,6))

axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, np.sin(x)*10000, label=r'$\cos( x ) * 1,000$')
axes.plot(x, x**2, label=r'$x^2$')
axes.plot(x, x**3, label=r'$x^3$')
axes.legend(loc='best', fontsize='large')

plt.show()
```

### Set limits to zoom in.

```python
x = np.linspace(0, 20, 1000)

fig = plt.figure(figsize=(10,6))

axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, np.sin(x)*10000, label=r'$\cos( x ) * 1,000$')
axes.plot(x, x**2, label=r'$x^2$')
axes.plot(x, x**3, label=r'$x^3$')
axes.legend(loc='best', fontsize='large')

axes2 = fig.add_axes([0.8, 0.8, 0.2, 0.2])
axes2.plot(x, np.sin(x)*10000, label=r'$\cos( x ) * 1,000$')
axes2.plot(x, x**2, label=r'$x^2$')
axes2.plot(x, x**3, label=r'$x^3$')
axes2.set_xlim(10, 11)

plt.show()
```

### Make a scatter plot.

```python
normx = np.random.randn(500)
normy = np.random.randn(500)

unix = np.random.uniform(0, 1, 500)
uniy = np.random.uniform(0, 1, 500)

fx = np.random.f(1, 50, 500)
fy = np.random.f(1, 50, 500)

geox = np.random.poisson(size=500)
geoy = np.random.poisson(size=500)

fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0,0].scatter(normx, normy, color='red')
axes[0,0].set_title('Normal Distribution')

axes[0,1].scatter(unix, uniy, color='purple')
axes[0,1].set_title('Uniform Distribution')

axes[1,0].scatter(fx, fy, color='blue')
axes[1,0].set_title(r'$f$ Distribution')

axes[1,1].scatter(geox, geoy, color='orange')
axes[1,1].set_title('Geometric Distribution')

for row in axes:
    for cell in row:
        cell.set_axis_off()

plt.show()
```

### Make a histogram.

```python
ndist = np.random.randn(10000)

fig, axes = plt.subplots()

axes.hist(ndist, 20)

plt.show()
```
