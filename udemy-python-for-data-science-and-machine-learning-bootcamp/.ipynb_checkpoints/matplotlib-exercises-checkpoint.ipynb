# Matplotlib Exercises

### Do the necessary imports.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Create the data to be used for these exercises. 

```python
x = np.arange(100)
y = x*2
z = x**2
```

# Exercise 1

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, y)
axes.set_title('Awesome Title')
axes.set_xlabel('x-axis')
axes.set_ylabel('y-axis')

plt.show()
```

# Exercise 2

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

axes1.plot(x, y)
axes2.plot(x, y)

plt.show()
```

# Exercise 3

```python
fig = plt.figure()

axes1 = fig.add_axes([0.0, 0.0, 1.0, 1.0])
axes1.plot(x, z)
axes1.set(xlabel='x', ylabel='z')

axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])
axes2.plot(x, y)
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_title('Zoom')
axes2.set_xlim(20, 22)
axes2.set_ylim(30, 50)
```

# Exercise 4

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 2))

axes[0].plot(x, y, linestyle='--', linewidth=4)
axes[0].set(xlabel='x', ylabel='y')

axes[1].plot(x, z, color='red', linewidth=4)
axes[1].set(xlabel='x', ylabel='z')

plt.tight_layout()
```
