# Seaborn Distribution Plots

```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

# Load a built-in data set.

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

# `seaborn.distplot()`

This will produce a nice histogram of any 1-dimensional array.

```python
sns.distplot(tips['total_bill'])
```

By default Seaborn includes a line for the kernel density estimation. There's an argument to remove that.

```python
sns.distplot(tips['total_bill'],
             kde=False)
```

One can also set the number of bins easily.

```python
sns.distplot(tips['total_bill'],
             kde=False,
             bins=5)
```

# `seaborn.jointplot()`

This is a method to show bivariate relationships, along with cool histograms on the outside. The `kind` argument offers several options:

* scatter (default)
* reg
* resid
* kde
* hex

```python
sns.jointplot(x=tips['total_bill'],
              y=tips['tip'])
```

```python
# Regression plot.
sns.jointplot(x=tips['total_bill'],
              y=tips['tip'],
              kind='reg')
```

```python
# Residual plot.
sns.jointplot(x=tips['total_bill'],
              y=tips['tip'],
              kind='resid')
```

```python
# Kernel density estimation plot.
sns.jointplot(x=tips['total_bill'],
              y=tips['tip'],
              kind='kde')
```

```python
# Cool hex plot. Useful though?
sns.jointplot(x=tips['total_bill'],
              y=tips['tip'],
              kind='hex')
```

# `seaborn.pairplot()`

The Seaborn equivalent of throwing everything at the wall and seeing what sticks. Probably more cool than interesting, but what do I know?

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,
             hue='sex')
```

# `seaborn.rugplot()`

The most interesting thing about rugplots is that they are the building blocks of kde plots.

```python
sns.rugplot(tips['total_bill'])
```

# Categorical Plots

Seaborn comes with a nice selection of plots for showing categorical data.

* factorplot
* boxplot
* violinplot
* stripplot
* swarmplot
* barplot
* countplot

# `sns.barplot()`

The classic plot that needs no introduction.

```python
sns.barplot(x=tips['sex'],
            y=tips['total_bill'])
```

By default the estimator object is the mean. Another function can be passed if desired, such as standard deviation.

```python
sns.set()
sns.barplot(x=tips['sex'],
            y=tips['total_bill'],
            estimator=np.std)
```
