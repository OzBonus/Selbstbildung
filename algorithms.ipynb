# Sieve of Eratosthenes

```python
%time
import math

def sieve(maximum):
    """Return a list of primes below the maximum"""
    numbers = [n for n in range(2, maximum+1)] # Numbers of interest.
    max_factor = round(math.sqrt(maximum))     # Highest factor to check.
    for n in numbers:
        if n == False:
            pass
        elif n <= max_factor:
            for idx in range(numbers.index(n), len(numbers), n):
                if numbers[idx] % n == 0 and numbers[idx] != n:
                    numbers[idx] = False
        else:
            return [n for n in numbers if n]

print(len(sieve(10000000)))
```

```python
%time
def sieve2(n):
    """
    Calculates prime numbers till a number n
    :param n: Number upto which to calculate primes
    :return: A boolean list contaning only primes
    """
    primes = [True] * (n+1)  # Setting all as True initially
    primes[0] = primes[1] = False  # Since 0 and 1 are not primes
    sqrt_n = math.ceil(math.sqrt(n))
    for i in range(2, sqrt_n, 1):
        if primes[i]:
            for j in range(2*i, n+1, i):
                primes[j] = False
    return primes

len(sieve2(10000))
```
